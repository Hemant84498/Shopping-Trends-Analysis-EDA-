# 🛍️ Shopping Trends Analysis

**Dataset:** Consumer Shopping Trends (3,900 customer records, 19 features)  
**Tools:** Python · pandas · matplotlib · seaborn · numpy  

**Objective:** Analyse customer shopping behaviour — including purchase patterns, demographics, seasonal trends, payment preferences, and discount impact — to surface actionable retail insights.

---
### 📋 Key Features in the Dataset
| Feature Group | Columns |
|---|---|
| Demographics | Age, Gender, Location |
| Purchase Details | Item Purchased, Category, Purchase Amount (USD), Season |
| Behaviour | Frequency of Purchases, Previous Purchases, Subscription Status |
| Transaction | Payment Method, Shipping Type, Discount Applied, Promo Code Used, Review Rating |

```python
# Importing required libraries for data analysis and visualization
# pandas -> data manipulation
# numpy -> numerical operations
# matplotlib & seaborn -> data visualization

import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns 
```

```python
# Loading the dataset into a pandas DataFrame
# The dataset contains shopping behavior and customer trends

df = pd.read_csv("shopping_trends.csv")
```

---
## 2. 📊 Data Visualisation & Insights
Exploring key trends through charts and grouped aggregations to understand customer behaviour and business performance.

```python
# plotting the histogram for check age groups
df["Age"].plot(kind = "hist")
plt.title("Histogram Of Age")
plt.show()
```

```python
# Show the all columns used in DataSet  
df.columns
```

```python
# Plotting the bar plot for check the no of counts Male vs Female 

df["Gender"].value_counts().plot(kind = "bar")
plt.title("Bar plot of Gender Distribution")
plt.show()
```

```python
# Identifying the 2 lowest-performing categories by average purchase amount
# sort_values(ascending=False) orders from highest to lowest, tail(2) picks the bottom 2

df.groupby("Category")["Purchase Amount (USD)"].mean().sort_values(ascending = False ).tail(2)
```

```python
# Sort the values from "Review Rating Column" In Ascending order 

df.sort_values("Review Rating" , ascending = False , inplace = True )
```

```python
# Displaying the full DataFrame after sorting by Review Rating in descending order
# Allows us to browse records from highest-rated to lowest-rated purchases

df
```

```python
# Plot pie chart for getting Subscription Status 

df["Subscription Status"].value_counts().plot( kind = "pie" )
plt.title("Pie Chart of Suscription")
plt.show()
```

```python
# Pie chart showing the top 5 most popular product colors among customers
# Helps understand color preferences for inventory and marketing decisions

df["Color"].value_counts().head().plot(kind = "pie")
plt.title("Top 5 Most Popular Colors")
plt.ylabel("")
plt.show()
```

```python
# Scatter plot exploring the relationship between Previous Purchases and Review Rating
# Using top 10 rows for a cleaner visualization
# Insight: Does buying more frequently lead to higher or lower satisfaction?

df.head(10).plot( x = "Previous Purchases" , y = "Review Rating" , kind = "scatter" )
plt.title("Scatter Plot - Previous Purchases vs Review Rating")
plt.xlabel("Previous Purchases")
plt.ylabel("Review Rating")
plt.show()
```

```python
# Total purchase revenue grouped by Season
# Identifies which season generates the highest overall sales

df.groupby("Season")["Purchase Amount (USD)"].sum()
```

```python
# Box plot: distribution of Purchase Amount across different purchase frequency groups
# Helps answer: do customers who shop more often spend more per transaction?

df.boxplot(column = "Purchase Amount (USD)" , by = "Frequency of Purchases")
plt.title("Purchase Amount by Frequency of Purchases")
plt.suptitle("")
plt.xlabel("Frequency of Purchases")
plt.ylabel("Purchase Amount (USD)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar chart of the top 5 preferred payment methods
# Reveals which payment options are most favored — useful for checkout UX decisions

df["Preferred Payment Method"].value_counts().sort_values(ascending = False).head().plot( kind = "bar")
plt.title("Top 5 Preferred Payment Methods")
plt.xlabel("Payment Method")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# Pivot table: Average Purchase Amount (USD) by Location and Item Purchased
# Rows = US States (Location), Columns = Product Items
# fill_value=0 replaces NaN for location-item combinations with no sales data
# Insight: Which products are most popular (by avg spend) in each state?

pd.pivot_table( values = "Purchase Amount (USD)" , index = "Location" , columns = "Item Purchased" , aggfunc = "mean" , fill_value = 0 , data = df)
```

```python
# Total revenue per payment method used at transaction time
# Complements the 'Preferred' payment method chart — shows actual usage vs preference

df.groupby("Payment Method")["Purchase Amount (USD)"].sum()
```

```python
# Average Review Rating grouped by whether a discount was applied
# Key question: Are discounted purchases rated higher or lower by customers?

df.groupby("Discount Applied")["Review Rating"].mean()
```

```python
# Number of purchases per Category, broken down by product Color
# Reveals color-category trends (e.g., which colors are most bought in Clothing vs Footwear)

df.groupby("Color")["Category"].value_counts()
```

```python
# Median number of previous purchases for each shopping frequency group
# Helps identify loyalty patterns — do more frequent shoppers have a longer purchase history?

df.groupby("Frequency of Purchases")["Previous Purchases"].median()
```

```python
# Average review rating per season
# Identifies if customer satisfaction varies seasonally — potential quality or fulfillment issues

df.groupby("Season")["Review Rating"].mean()
```

```python
# Bar chart: count of purchases per season
# Shows which seasons are peak shopping periods

df["Season"].value_counts().plot(kind = "bar")
plt.xlabel("Season")
plt.ylabel("Count")
plt.title("Count of Purchases per Season")
plt.show()
```

```python
# Pie chart: proportion of total purchases across product categories
# Shows the share of sales for Clothing, Accessories, Footwear, and Outerwear

df["Category"].value_counts().plot( kind = "pie")
plt.title("Distribution of Purchases by Category")
plt.ylabel("")
plt.show()
```

```python
# Box plot comparing Review Rating distributions between Male and Female customers
# Checks if there's a gender-based difference in customer satisfaction

df.boxplot(column = "Review Rating" , by = "Gender")
plt.xlabel("Gender")
plt.ylabel("Review Rating")
plt.title("Review Rating Distribution by Gender")
plt.suptitle("")
plt.show()
```

```python
# Histogram showing the frequency distribution of Purchase Amount (USD)
# 10 bins helps visualize how purchases are spread across price ranges
# A fairly uniform distribution would suggest no strong price preference

df["Purchase Amount (USD)"].plot(kind = "hist" , bins = 10)
plt.xlabel("USD")
plt.ylabel("Frequency")
plt.title("Histogram of Purchase Amount Distribution")
plt.show()
```

```python
# Bar chart: average review rating for the top 5 product colors
# Explores whether specific colors receive consistently better or worse ratings

df.groupby("Color")["Review Rating"].mean().head().plot(kind = "bar")
plt.title("Mean Review Rating by Color (Top 5)")
plt.xlabel("Color")
plt.ylabel("Mean Review Rating")
plt.show()
```

```python
# Pie chart: share of total revenue contributed by each payment method
# Visualizes the revenue distribution across payment channels

df.groupby("Payment Method")["Purchase Amount (USD)"].sum().plot(kind = "pie")
plt.title("Revenue Share by Payment Method")
plt.ylabel("")
plt.show()
```

```python
# Box plot: Purchase Amount vs Frequency of Purchases (enhanced labels)
# Visualizes spending spread for each frequency category to detect high/low spenders

df.boxplot(column = "Purchase Amount (USD)" , by = "Frequency of Purchases")
plt.xlabel("Frequency Of Purchases")
plt.ylabel("Purchase Amount (USD)")
plt.title("Box Plot: Purchase Amount by Frequency of Purchases")
plt.suptitle("")
plt.xticks(rotation=45)
plt.show()
```

```python
# Pie chart: total purchase revenue share by Season
# Provides a quick visual of which season contributes most to overall revenue

df.groupby("Season")["Purchase Amount (USD)"].sum().plot(kind = "pie")
plt.title("Revenue Share by Season")
plt.ylabel("")
plt.show()
```
